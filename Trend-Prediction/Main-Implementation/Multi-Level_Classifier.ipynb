## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from skmultilearn.problem_transform import ClassifierChain
from sklearn.model_selection import GridSearchCV
import numpy as np
# define model
params = {"classifier__C": np.logspace(-3,3,7), "classifier__penalty": ["l2"]} # l1 lasso l2 ridge

logreg = ClassifierChain(LogisticRegression())
logreg_grid = GridSearchCV(logreg, params, cv=2, verbose=2)
# define search
result = logreg_grid.fit(X_train, y_train)
# summarize result
print('Best Score: %s' % result.best_score_)
print('Best Hyperparameters: %s' % result.best_params_)
```

```python
grid_pred = logreg_grid.predict_proba(X_test)
dense_array = grid_pred.toarray()

grid_pred1 = []
for arr in dense_array:
    grid_pred1.append(np.where(arr > 0.5, 1, 0))

print(grid_pred1)
# accuracy
print("Accuracy = ",accuracy_score(y_test, grid_pred1))
```

```python
cr = classification_report(y_test, grid_pred1)
print(cr)
write_log("LR-CC", cr)
```

```python
import os
pickle = pkl.dump(logreg_grid, open(os.path.join(pkl_file, 'lr_grid_classchain.pickle'), 'wb'))
```

## Decision Tree

```python
from sklearn import decomposition, datasets
from sklearn import tree
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
std_slc = StandardScaler()
pca = decomposition.PCA()
dec_tree = tree.DecisionTreeClassifier()
pipe = Pipeline(steps=[('std_slc', std_slc),
                           ('pca', pca),
                           ('dec_tree', dec_tree)])
```

```python
n_components = list(range(1, X_train.shape[1]+1, 1))
criterion = ['gini', 'entropy']
max_depth = [2,4,6,8,10,12]
parameters = dict(pca__n_components=n_components,
                      dec_tree__criterion=criterion,
                      dec_tree__max_depth=max_depth)
```

```python
clf_GS = GridSearchCV(pipe, parameters)
clf_GS.fit(X_train, y_train)
dtPreds = clf_GS.predict(X_test)
```

```python
cr = classification_report(y_test, dtPreds)
print(cr) 
write_log("DecisionTree-Grid", cr)
```

```python
pickle = pkl.dump(clf_GS, open(os.path.join(pkl_file, 'decision_tree_grid.pickle'), 'wb'))
```
